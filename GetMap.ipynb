```python
import numpy as np

row=12
col=12
road=int(row*col/2.5)
honey=5
hyperlink=3

def makeWall(graph):   
     for i in range(row):         
         graph.append([])
         for j in range(col):
             graph[i].append('x')
             
     rowRandom=np.random.randint(1,row-1)
     colRandom=np.random.randint(1,col-1)
     graph[rowRandom][colRandom]='b'
     
     return graph
    
def makeRoad(graph,road):
    while road>0:
        rowRandom=np.random.randint(1,row-1)
        colRandom=np.random.randint(1,col-1)
        if graph[rowRandom][colRandom]=='x':
            graph[rowRandom][colRandom]='.'
            road-=1
    return graph

def makeHoney(graph,honey):
    while honey>0:
        rowRandom=np.random.randint(1,row-1)
        colRandom=np.random.randint(1,col-1)
        if graph[rowRandom][colRandom]=='x':
            graph[rowRandom][colRandom]='h'
            honey-=1
    return graph

def makeHyperlink(graph,hyperlink):
    H=hyperlink*2
    num=1
    while H>0:
        rowRandom=np.random.randint(1,row-1)
        colRandom=np.random.randint(1,col-1)
        if graph[rowRandom][colRandom]=='x':
            graph[rowRandom][colRandom]=num
            if H%2:
                num+=1
            H-=1           
            
    return graph

def printGraph(graph):
     string=""
     for i in range(row):
         for j in range(col):
             string+=str(graph[i][j])+" "
         string+="\n"   
     print(string)
     return string

    
graph=[]
_graph=makeWall(graph)
_graph=makeRoad(_graph,road)
_graph=makeHoney(_graph,honey)
_graph=makeHyperlink(_graph,hyperlink)
_string=printGraph(_graph)

saveFile=open('input4.txt','w')
saveFile.write(_string)
saveFile.close()

```

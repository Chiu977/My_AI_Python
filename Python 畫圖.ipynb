```python
x = y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax = plt.gca(projection='3d')
ax.plot_surface(X, Y, Z)
```

```python
c = 2
a = 1.2

t = np.linspace(0, 2*np.pi, 2000)
u, v = np.meshgrid(t, t)
x = (c + a*np.cos(v)) * np.cos(u)
y = (c + a*np.cos(v)) * np.sin(u)
z = a * np.sin(v)

ax = plt.gca(projection='3d')

ax.plot_surface(x, y, z)
```

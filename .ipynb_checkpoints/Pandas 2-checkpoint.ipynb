```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
df=pd.read_csv('http://stats.moe.gov.tw/files/detail/107/107_student.csv')
```

```python
df.head()
```

```python
colname=df.loc[1].values
```

```python
df=df[2:]
df.head()
```

```python
df.columns=colname
```

```python
df.head()
```

```python
undergraduate=df[(df['日間∕進修別']=='D 日') & (df['等級別']=='B 學士')]
undergraduate.head()
```

```python
df2=undergraduate.loc[:,'學校名稱':'女生計'].copy()
```

```python
df2.head()
```

```python
import locale
```

```python
locale.setlocale(locale.LC_NUMERIC,'')
```

```python
df2['男生計']=df2['男生計'].apply(locale.atof)
```

```python
df2['男生計']=df2['男生計'].apply(locale.atof)df2['女生計']=df2['女生計'].apply(locale.atof)
```

```python
df2['女男比例']=df2.女生計/df2.男生計
```

```python
df2.head()
```

```python
df3=df2.sort_values(by=['女男比例'],ascending=False)
```

```python
df3[:20].plot.bar(x='學校名稱',y=['女男比例'])
```

# 1. import需要的工具

```python
%env KERAS_BACKEND=tensorflow
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact_manual
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
from keras.datasets import mnist
from keras.utils import to_categorical
```

# 2. 讀取輸入資料並做處理

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.reshape(60000, 28*28)
x_test = x_test.reshape(10000, 28*28)
```

```python
x_train = (x_train - x_train.min())/(x_train.max() - x_train.min())
x_test = (x_test - x_test.min())/(x_test.max() - x_test.min())
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

# 3.import Model 來使用Function API

```python
from keras.models import Model
from keras.layers import Input
from keras.layers import concatenate, add
```

# 4. 定義神經網路模型

### 神經網路的actication使用效果較佳的relu，模型如下:



```python
x = Input(shape=(784,))
```

## 定義函數

```python
f_1 = Dense(50, activation='relu')
f_2 = Dense(50, activation='relu')

f_3 = Dense(100, activation='relu')
f_4 = Dense(100, activation='relu')

f_5 = Dense(500, activation='relu')

f_6 = Dense(20, activation='relu')
f_7 = Dense(20, activation='relu')
f_8 = Dense(20, activation='relu')
f_9 = Dense(20, activation='relu')

f_10 = Dense(10, activation='softmax')
```

## 定義變數

```python
h_11=f_1(x)
h_12=f_2(x)

h_21=f_3(h_11)
h_22=f_4(h_12)
u = concatenate([h_21, h_22])

h_3=f_5(u)

h_41=f_6(h_3)
h_42=f_7(h_3)
h_43=f_8(h_3)
h_44=f_9(h_3)

uu=concatenate([h_41, h_42,h_43,h_44])
y=f_10(uu)
```

## 確認模型的架構

```python
model = Model(x, y)
model.summary()
```

# 5. 開始訓練模型，loss fuction使用mse、optimizer使用SGD (Learnig rate=0.087)

```python
model.compile(loss='mse', optimizer=SGD(lr=0.087), metrics=['accuracy'])
```

```python
model_out=model.fit(x_train, y_train, batch_size=100, epochs=50,verbose =1,validation_data = (x_test,y_test))
```

```python
model.compile(loss='mse', optimizer=SGD(lr=0.1), metrics=['accuracy'])
score_Com = model.evaluate(x_test, y_test, batch_size=1000)
print("Loss: %f" %score_Com[0])
print("準確率: %f" %(score_Com[1]*100))
```

# 6. 把數據圖像化

```python
plt.plot(model_out.history["acc"])

plt.title("model training accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["Relu"],loc="best")
plt.show()
```

```python
plt.plot(model_out.history["val_loss"])

plt.title("model validation loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["Relu"],loc="best")
plt.show()
```

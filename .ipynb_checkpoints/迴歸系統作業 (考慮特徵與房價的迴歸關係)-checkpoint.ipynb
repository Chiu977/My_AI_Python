```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
```

# **1. 讀取檔案以及確認null**

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.isnull().any()
```

# **2. 選取特徵**

<從train.csv的81筆資料去做選取較合理的特徵以及採用數字型態的特徵來做分析>

1.LotArea  ->地塊面積（平方英尺）

2.GrLivArea ->生活面積（平方英尺）

3.GarageArea ->平方英尺車庫大小

4.YearBuilt ->原始施工日期

5.df.YrSold  ->年銷售

# **3. 預測目標**
SalePrice ->房屋的銷售價格

```python
#X_dataset=df[['MSSubClass','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold']]
X_dataset=df[['LotArea','GarageArea','YearBuilt','YrSold','GrLivArea']]
X_dataset.head()
```

```python
Y_dataset=df[['SalePrice']]
Y_dataset.head()
```

# **4. 選取的特徵轉製成矩陣**

```python
LotArea_arr=np.array(df.LotArea)
GrLivArea_arr=np.array(df.GrLivArea)
GarageArea_arr=np.array(df.GarageArea)
YearBuilt_arr=np.array(df.YearBuilt)
YrSold_arr=np.array(df.YrSold)
SalePrice_arr=np.array(df.SalePrice)
```

# **5. LotArea對SalePrice的迴歸分析**

```python
X_Lot = LotArea_arr.reshape(len(LotArea_arr), 1)

regr_Lot=LinearRegression()
regr_Lot.fit(X_Lot,SalePrice_arr)
Y_Lot=regr_Lot.predict(X_Lot)

plt.figure(dpi=150)
plt.scatter(X_Lot,SalePrice_arr) #LotArea 對 SalePrice的分布
plt.plot(X_Lot,Y_Lot,'r')        #LotArea 對 SalePrice的迴歸線

plt.xlabel("LotArea")
plt.ylabel("SalePrice")
```

### ★ 數據裡推測有涵蓋離群值

# **6. GrLivArea對SalePrice的迴歸分析**

```python
X_GrL = GrLivArea_arr.reshape(len(GrLivArea_arr), 1)

regr_GrL=LinearRegression()
regr_GrL.fit(X_GrL,SalePrice_arr)
Y_pred=regr_GrL.predict(X_GrL)

plt.figure(dpi=150)
plt.scatter(X_GrL,SalePrice_arr)    #GrLivArea 對 SalePrice的分布
plt.plot(X_GrL,Y_pred,'r')          #GrLivArea 對 SalePrice的迴歸線

plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
```

### ★ 有成明顯的線性關係

# **7. GarageArea對SalePrice的迴歸分析**

```python
X_Gar = GarageArea_arr.reshape(len(GarageArea_arr), 1)

regr_Gar=LinearRegression()
regr_Gar.fit(X_Gar,SalePrice_arr)
Y_Gar=regr_Gar.predict(X_Gar)

plt.figure(dpi=150)
plt.scatter(X_Gar,SalePrice_arr)   #GarageArea 對 SalePrice的分布
plt.plot(X_Gar,Y_Gar,'r')          #GarageArea 對 SalePrice的迴歸線

plt.xlabel("GarageArea")
plt.ylabel("SalePrice")
```

# **8. YrSold對SalePrice的迴歸分析**

```python
X_Yr = YrSold_arr.reshape(len(YrSold_arr), 1)

regr_Yr=LinearRegression()
regr_Yr.fit(X_Yr,SalePrice_arr)
Y_Yr=regr_Yr.predict(X_Yr)

plt.figure(dpi=150)
plt.scatter(X_Yr,SalePrice_arr)    #YrSold 對 SalePrice的分布
plt.plot(X_Yr,Y_Yr,'r')            #YrSold 對 SalePrice的分布

plt.xlabel("YrSold")
plt.ylabel("SalePrice")
```

### ★ 為離散數據

# **9. YearBuilt對SalePrice的迴歸分析**

```python
X_Year = YearBuilt_arr.reshape(len(YearBuilt_arr), 1)

regr_Year=LinearRegression()
regr_Year.fit(X_Year,SalePrice_arr)
Y_Year=regr_Year.predict(X_Year)

plt.figure(dpi=150)
plt.scatter(X_Year,SalePrice_arr)    #YearBuilt 對 SalePrice的分布
plt.plot(X_Year,Y_Year,'r')          #YearBuilt 對 SalePrice的迴歸線

plt.xlabel("YearBuilt")
plt.ylabel("SalePrice")
```

# **10. 五個特徵對SalePrice的正確率計算**

```python
x_train, x_test, y_train, y_test = train_test_split(X_dataset,Y_dataset, test_size=0.3, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
#Y_pred=model.predict(x_test)
model.score(x_test,y_test)
```

### ★ 71.6%正確率蠻不錯的

###    以下為兩個評估分析，第一個指選取LotArea、GarageArea、YearBuilt、YrSold ，發現明顯的下降。如果把LotArea(有離群值)和YrSold(為離散分布)的去計算正確率為71.4%，沒有明顯的下降。綜合以上結論，能推斷GrLivArea為強特徵，所以和SalePrice有明顯的線性關係。

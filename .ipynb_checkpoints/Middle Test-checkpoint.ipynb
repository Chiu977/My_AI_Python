```python
import numpy as np
from enum import Enum
import copy
import time
```

## 1.Enum

```python
class NOROAD(Enum):
    #NoFINISH
    NO=0
    FINISH=1
    
class Direction(Enum):
    UP = 0
    DOWN = 1
    RIGHT_UP = 2
    LEFT_UP = 3
    RIGHT_DOWN = 4
    LEFT_DOWN = 5
    
class PointType(Enum):
    ROAD = 0
    WALL = 1
    HYPERLINK = 2
```

## 2.

#### 1.Location:rowcolumrowcolum__eq____ne__getDirection
#### 2.RoadHoneyWall:LocationType
#### 3.Bee: LocationsetLocation
#### 4.Hyperlink:Location
#### 5.Hive: txt

```python
class Location(object):
    def __init__(self,row,col):
        self.row=row
        self.col=col        
    def __eq__(self, other):
        return self.row == other.row and self.col == other.col
    def __ne__(self, other):
        return self.row != other.row or self.col != other.col     
    def getDirection(self,direct):
        loc=Location(self.row,self.col)
        if direct==Direction.UP:
            loc.row-=1
        elif direct==Direction.DOWN:
            loc.row+=1
        elif direct==Direction.RIGHT_UP:
            loc.col+=1
        elif direct==Direction.RIGHT_DOWN:
            loc.col+=1
            loc.row+=1
        elif direct==Direction.LEFT_UP:
            loc.col-=1
            loc.row-=1
        elif direct==Direction.LEFT_DOWN:
            loc.col-=1
        return loc
    
class Road(Location):
    def __init__(self, row, col):
        super().__init__(row, col)
        self.type=PointType.ROAD
    def __str__(self):
        return ". "

class Honey(Location):
    def __init__(self, row, col):
        super().__init__(row, col)
        self.type=PointType.ROAD
    def __str__(self):
        return "h "
    
class Wall(Location):
    def __init__(self, row, col):
        super().__init__(row, col)
        self.type=PointType.WALL
    def __str__(self):
        return "x "
    
class Hyperlink(Location):
    def __init__(self, row, col, tid):
        super().__init__(row, col)
        self.hys=Location(-1, -1)
        self.id=int(tid)
        self.type=PointType.HYPERLINK
    def getHys(self):
        return self.hys
    def setHys(self, row, col): #Hyperlink
        self.hys.row=row
        self.hys.col=col
    def __str__(self):
        return str(self.id)+" "

class Bee(Location):
    def __init__(self, row, col):
        super().__init__(row, col)
        self.type=PointType.ROAD
    def setLocation(self, row, col):
        self.loc=Location(row, col)        
    def __str__(self):
        return "b "
    
class Hive(object):
    def __init__(self,Graph,size):
        self.gRow=size[0]
        self.gCol=size[1]
        self.bee=Bee(0,0) #
        self.honey=[] #
        hyperlink=[] #
        self.graph=[] #
        for i in range (self.gRow):
            self.graph.append([])
            for j in range (self.gCol):
                if Graph[i][j]=='x':
                    self.graph[i].append(Wall(i,j)) #Wall
                elif Graph[i][j]=='.':
                    self.graph[i].append(Road(i,j)) #Road
                elif Graph[i][j]=='b':
                    self.bee.setLocation(i,j) #bee
                    self.graph[i].append(Bee(i,j)) #Bee
                elif Graph[i][j]=='h':
                    self.graph[i].append(Honey(i,j))  #Honey
                    self.honey.append(Honey(i,j))  #Honeyhoney
                else:
                    self.graph[i].append(Hyperlink(i,j,Graph[i][j]))
                    flag=False
                    for k in range(len(hyperlink)):  
                         if hyperlink[k]['id']==Graph[i][j]:  #hyperlinkid
                            self.graph[i][j].setHys(hyperlink[k]['row'],hyperlink[k]['col']) #Hyperlink
                            self.graph[hyperlink[k]['row']][hyperlink[k]['col']].setHys(i,j)
                            del hyperlink[k]
                            flag=True
                            break
                    if not flag:
                        hyperlink.append({'id':Graph[i][j],'row':i,'col':j})   #HyperlinkhyperlinkLibrary
    def __str__(self): #graph
        string=""
        for i in range(self.gRow):
             for j in range(self.gCol):
                    string+=str(self.graph[i][j])
             string+="\n"   
        return string
    def getgraph(self,start): #graph
        return self.graph[start.row][start.col]
    
    def HoneyCheck(self,start): #honeyHoney
        for honey in self.honey:
            if honey.row == start.row and honey.col==start.col:
                return True
        return False
    
    def TakeHoney(self,start): #honeyHoney
        for honey in self.honey:
            if honey.row == start.row and honey.col==start.col:
                self.honey.remove(honey)
                
    def getHyperLink(self,start): #TypeHYPERLINK
        if self.graph[start.row][start.col].type==PointType.HYPERLINK:
            return self.graph[start.row][start.col].getHys()
        return start
```

## 3.()()

```python
def readFile(filename):
    f=open(filename,'r')
    elements=[]
    for line in f:
        line=line.split()
        elements.append(line)
    ele_array = np.array(elements)
    return ele_array, ele_array.shape
```

## 4. Search_IDStime

```python
def Search_IDS(hive,maxDepth):
    StartTime=time.time()
    route,cost=IDS(hive,maxDepth) #
    EndTime=time.time()
    print('time=', EndTime-StartTime, 'sec')
    print('cost: ',cost)
    string=[]
    for i in range(len(route)-1,0,-1): #for
        
        if route[i]==Direction.UP:
            string.append('')
        elif route[i]==Direction.DOWN:
            string.append('')
        elif route[i]==Direction.RIGHT_UP:
            string.append('')
        elif route[i]==Direction.RIGHT_DOWN:
            string.append('')
        elif route[i]==Direction.LEFT_UP:
            string.append('')
        elif route[i]==Direction.LEFT_DOWN:
            string.append('')
    
    print(string)
```

```python
def IDS(hive,maxDepth):
    start=hive.bee.loc #hivebee
    for lteration in range (maxDepth): #
        pathlist=[]
        costlist=[]   
        for direct in list(Direction): #
            Nhive=copy.deepcopy(hive)
            path,cost=DLS(Nhive,initWalk(hive,hive.bee.loc),start.getDirection(direct),lteration) 
            costlist.append(cost) #()
            pathlist.append(path)
        Direction_index,way = Mincost(costlist) #()
        path =pathlist[Direction_index]  #path
        path.append(Direction(Direction_index)) 
        if path[0]==NOROAD.FINISH: #FINISHcost
            return path, way+1
    print('NO found')    
    return path, way+1
```

```python
def DLS(Nhive,walk,start,lteration):
    if lteration<=0: #0
        return [NOROAD.NO],int(infinite)
    start=Nhive.getHyperLink(start) #
    if Nhive.getgraph(start).type==PointType.WALL or walk[start.row][start.col]!=0: #typeWALL
        return [NOROAD.NO],int(infinite)
    walk[start.row][start.col]=1 #
    if Nhive.HoneyCheck(start): #
        Nhive.TakeHoney(start)   #
        if len(Nhive.honey)>0: #
            walk=initWalk(Nhive,start)
        else:
            return [NOROAD.FINISH],0 #0
     
    #()
    new_walk=copy.deepcopy(walk)
    Costlist=[]
    Pathlist=[]
    for direct in list(Direction):
        N_hive=copy.deepcopy(Nhive)
        Path,Cost=DLS(N_hive,new_walk,start.getDirection(direct),lteration-1)
        Costlist.append(Cost)
        Pathlist.append(Path)
    Direction_index,way=Mincost(Costlist)
    path =Pathlist[Direction_index]
    path.append(Direction(Direction_index))
    return path, way+1
```

```python
def initWalk(hive,start):
    # 0  1
    unwalked=[]
    for i in range(hive.gRow):
        unwalked.append([])
        for j in range(hive.gCol):
            unwalked[i].append(0)  
    unwalked[start.row][start.col]=1 #
    return unwalked  

def Mincost(costlist): #
    min_index=np.argmin(np.asarray(costlist))
    return min_index,costlist[min_index]
```

```python
global infinite
infinite=1e10
maxDepth=50
file,size=readFile('input.txt') #
hive=Hive(file,size) #Hive
print(hive) #
Search_IDS(hive,maxDepth) #IDS
```
